```
import numpy as np
import pandas as pd
```

```
#1: Use Pandas package to create and display a DataFrame from a specified dictionary data. Please create 3 rows and 4 columns given in the table above..
seriesNameDict = {0: 'Tom',
                   1: 'Sally',
                   2: 'Alex'}
seriesNameDict = pd.Series(seriesNameDict)
seriesNameDict
seriesGenderDict = {0: 'M',
                   1: 'F',
                   2: 'M'}
seriesGenderDict = pd.Series(seriesGenderDict)
seriesGenderDict

seriesIncomeDict = {0: '100000',
                   1: '150000',
                   2: '200000'}
seriesIncomeDict = pd.Series(seriesIncomeDict)
seriesIncomeDict
seriesOccupationDict = {0: 'Researcher',
                   1: 'Manager',
                   2: 'Doctor'}
seriesOccupationDict = pd.Series(seriesOccupationDict)
seriesOccupationDict
personInfo = pd.DataFrame({'Name': seriesNameDict,
                       'Gender': seriesGenderDict,
                       'Income': seriesIncomeDict,
                           'Occupation': seriesOccupationDict})

print(personInfo)
```

```
#2: Display the summary of the data in the DataFrame.
personInfo.describe()
```

```
#3: Print only the first row of the DataFrame.
print(personInfo.iloc[0])
```

```
#4: Print first two columns from the DataFrame
first_two_columns = personInfo.iloc[:, :2]
print(first_two_columns)
```

```
#5: Replace the Income for Sally to 160000
personInfo['Income'] = personInfo['Income'].replace(['150000'], '160000')
print(personInfo)
```

```
#6: Select the rows where income is more than equal to 100000
personInfo['Income'] = personInfo['Income'].astype('int64')
selected_rows = personInfo[personInfo['Income'] >= 100000]
print(selected_rows)
```

```
#7: Insert a new row with values Sam ,F, 120000, Nurse
personInfo.loc[len(personInfo.index)] = ['Sam', 'F', 120000, 'Nurse' ]
print(personInfo)
```

```
#8: Print the row that has a maximum income value
df2=personInfo.query('Income == Income.max()')
print(df2)
```

```
#9: Print datatypes of each columns
print(personInfo.dtypes)
```

```
#10: Sort the data based on income
personInfo.sort_values(by='Income', ascending=True)
```
